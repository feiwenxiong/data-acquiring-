```python
import requests
from lxml import etree
import pandas as pd
import os

#####################################################
# 产生保存的路径
path = input('波妞能爬取豆瓣Top250电影信息哦，让波妞为主人服务哇：请输入主人要保存到的文件夹地址>>>>>>')


# 判断存在这个moveiposter文件夹否


#生成新的保存图片的文件夹路径
path_movieposter = path +  '\\' +   'movieposter'


# 判断存在这个moveiposter文件夹否,有就pass，没有就创建
if os.path.exists(path_movieposter):
    print('波妞先看看movieposter的文件夹是否存在:',os.path.exists(path_movieposter))
    pass
# os.chdir('path')
else:
    print('波妞先看看movieposter的文件夹是否存在:',os.path.exists(path_movieposter))
    os.mkdir(path_movieposter)
####################################################

MOVIES = []
IMGURLS = []


def get_html(url):

    headers = {'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/75.0.3770.100 Safari/537.36'}
    try:
        html = requests.get(url,headers = headers)
        html.encoding = html.apparent_encoding
        if html.status_code == 200:
            print('波妞成功获取源代码啦，开心')
            # print(html.text)
    except Exception as e:
        print('波妞获取源代码失败，呜呜呜:%s' % e)

    return html.text




def parse_html(html):

    movies = []
    imgurls = []
    html = etree.HTML(html)
    lis = html.xpath("//ol[@class = 'grid_view']/li")

    for li in lis:
        name = li.xpath(".//a/span[@class='title'][1]/text()")[0]
        director_actor = "".join(li.xpath(".//div[@class='bd']/p/text()[1]")[0].replace(' ','').replace('\n','').replace('/','').split())
        info = "".join(li.xpath(".//div[@class='bd']/p/text()[2]")[0].replace(' ','').replace('\n','').split())
        rating_score = li.xpath(".//span[@class='rating_num']/text()")[0]
        rating_num = li.xpath(".//div[@class='star']/span[4]/text()")[0]
        introduce = li.xpath(".//p[@class='quote']/span/text()")

        if introduce:
            movie = {'name': name, 'director_actor': director_actor, 'info': info, 'rating_score': rating_score,
                     'rating_num': rating_num, 'introduce': introduce[0]}
        else:
            movie = {'name': name, 'director_actor': director_actor, 'info': info, 'rating_score': rating_score,
                     'rating_num': rating_num, 'introduce': None}
        imgurl = li.xpath(".//img/@src")[0]

        movies.append(movie)
        imgurls.append(imgurl)

    return movies,imgurls



def download_img(url,movie):
    
#     path = r'C:\Users\richa\Desktop\电影top250'
    
#     if os.path.exists('movieposter'):
#         pass
#     else:
#         os.mkdir('movieposter')

# #     if 'movieposter'not in os.listdir(r'C:\Users\richa\Desktop\电影top250'):
# #         try:
# #             os.mkdir('movieposter')
# #         except FileNotFoundError:
# #             pass
# #     else:
# #         pass
    os.chdir(path_movieposter)

    img = requests.get(url).content

    with open(movie['name'] + '.jpg','wb') as f:
        print('波妞正在努力下载哦 ： %s' % url)
        f.write(img)

####################################################################################################################        

for i in range(10):
    url = 'https://movie.douban.com/top250?start=' + str(i * 25) + '&filter='

    html = get_html(url)
    movies = parse_html(html)[0]
    imgurls = parse_html(html)[1]

    MOVIES.extend(movies)
    IMGURLS.extend(imgurls)

for i in range(250):
    download_img(IMGURLS[i],MOVIES[i])

os.chdir(path)
moviedata = pd.DataFrame(MOVIES)
moviedata.to_csv('movie.csv')
print('哇哇哇！！！波妞开心地告诉主人：电影信息成功保存到本地啦！！！！！波妞波妞！！')
```
