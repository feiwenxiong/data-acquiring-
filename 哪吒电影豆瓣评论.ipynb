```python
# 产生保存的路径
path = input('波妞能爬哪吒额电影评论哦，让波妞为主人服务哇：请输入主人要保存到的文件夹地址>>>>>>')

#生成新的保存图片的文件夹路径path
path_final = path +  '\\' +   '哪吒评论'

# 判断存在这个moveiposter文件夹否,有就pass，没有就创建
if os.path.exists(path_final):
    print('波妞先看看\'哪吒评论\'文件夹是否存在:',os.path.exists(path_final))
    pass
# os.chdir('path')
else:
    print('波妞先看看\'哪吒评论\'的文件夹是否存在:',os.path.exists(path_final))
    os.mkdir(path_final)
```

```python
IMGURLS=[]
INFOS=[]

# 得到url后，得到html
def get_html(url,params):
    
    headers={'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/75.0.3770.100 Safari/537.36'
                }
    try:
        html = requests.get(url,headers= headers,params=params)
        html.encoding = 'utf-8'
        if html.status_code == 200:
            print('波妞收到宗介要的源代码了！！，嘻嘻')
    except Exception as e:
        print('波妞获取源代码失败，呜呜呜:%s' % e)
    
    return html.text
              
```

```python
def parse_html(html):
    
    infos=[]
    imgurls=[]

    html = etree.HTML(html)
    divs = html.xpath('//*[@class="comment-item"]')  
    
    
    for div in divs:
        name = div.xpath('.//a/@title')[0]
        
        imgurl = div.xpath('.//a/img/@src')[0]
        
        comment = div.xpath('.//p/span[@class ="short"]/text()')[0].replace('\n','')
        
        info ={'name':name,'comment':comment,'image':imgurl} # 封装成字典形式
        
        
        imgurls.append(imgurl)
        infos.append(info)
        
        
    return infos,imgurls
        
        
```

```python

def downloadimage(url,info):
    os.chdir(path_final)  # 在哪里产生文件夹
   
    img = requests.get(url).content  # 图片产生二进制数据
    
    # 产生文件 with open 
    
    try:
        
        with open(info['name'] +'_'+'.jpg','wb') as f:    # wb = = write binary 
            print('波妞正在努力下载哦 ： %s' % url)
            f.write(img)
    except Exception as e:
        pass

    
```

```python
# main 函数
for i in range(20):
    url='https://movie.douban.com/subject/26794435/comments?sort=new_score&status=P'
    params = {'start':20* i,'limit':20}
    
    html = get_html(url,params=params)
    infos = parse_html(html)[0]
    imgurls = parse_html(html)[1]
    
    
    INFOS.extend(infos)
    IMGURLS.extend(imgurls)

for i in range(20*20):
    downloadimage(IMGURLS[i],INFOS[i])
    
os.chdir(path)
infosdata = pd.DataFrame(INFOS,columns=['name','image','comment'])
infosdata.to_csv('infos.csv',encoding='utf_8_sig')
print('哇哇哇！！！波妞开心地告诉主人：评论成功保存到本地啦！！！！！波妞波妞！！')
```
